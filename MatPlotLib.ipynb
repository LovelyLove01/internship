# WHICH METHOD IS MORE IDEAL? Pyplot method VS Matplotlib OO Method(Object Oriented )
* When plotting something real quick, it's ok to use pyplot method#
* When plotting something more advanced use the OO method


```python
over_50= heart_disease[heart_disease["age"]>50]
over_50
```

```python
over_50.plot(kind="scatter", figsize=(10,6),
            x= "age",
            y="chol",
            c="target");
```

```python
#OO Method from scratch
fig, ax =plt.subplots(figsize=(10,6))
#plot the data
scatter=ax.scatter(x=over_50["age"],
                  y= over_50["chol"],
                  c= over_50["target"],
                  cmap="summer")
#Customize the plot
ax.set(title="Heart Disease and Cholesterol Level ",
      xlabel="Age",
      ylabel="Cholesterol Level")
#To add legend
ax.legend(*scatter.legend_elements(), title="Target")
#To indicate the average cholesterol
ax.axhline(over_50["chol"].mean(), linestyle="--");
```

```python
over_50.head()
```

```python
#Subplot of age, chol and thalach
fig, (ax0,ax1)=plt.subplots(nrows=2,ncols=1,figsize=(10,10), sharex=True)
scatter= ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                    c=over_50["target"],
                    cmap="summer")

scatter= ax1.scatter(x=over_50["age"],
                   y=over_50["thalach"],
                   c=over_50["target"],
                    cmap="winter")

ax0.set(title="Heart Disease and Cholesterol",
       xlabel="Age",
       ylabel="Cholesterol")
ax1.set(title="Heart Disease and Max Heart Rate",
       xlabel="Age",
       ylabel="Thalach");


ax0.legend(*scatter.legend_elements(), title="Target")
ax1.legend(*scatter.legend_elements(), title="Target")

ax0.axhline(over_50["chol"].mean(), linestyle="--")
ax1.axhline(over_50["thalach"].mean(), linestyle="--");

#To use just one scale label for both axes, input share=True

#To add title to the figure

fig.suptitle("Heart Disease Analysis", fontsize=18, fontweight="bold");
```
